## return
  - return의 정의
     - 함수 내부의 실행을 종료하고, 호출한 곳으로 결과값을 전달하는 키워드
  - return의 핵심 역할 2가지  
    - 값 반환: 함수에서 계산되거나 처리된 데이터(결과물)를 외부로 내보냄  
    - 함수 종료: return문이 실행되는 즉시 뒤에 코드가 더 있더라도 함수가 끝남
  - 아무것도 리턴하지  않으면 None을 출력함.  None은 아무것도 없다는 의미임.
  - 함수에서 return을 2번 이상 사용할 수 없다.
  - 함수의 결과를 2개 이상으로 받고 싶다면 unpacking해서 받는다.


````python
def proc(a):
    print(a)
    return # 함수 중간에 빠져나올떄

#아무것도 리턴하지 않으면 none을 출력함. none은 아무것도 없다는 의미
print(proc(3))



````

````python
#proc 함수에서 합계랑 평균을 구해서 리턴하는 프로그램을 구현

def proc(*a):
    total = 0
    avg = 0

    total = sum(a)
    avg = total / len(a)
    return total, avg

total,avg = proc(1,2,3,5,6,7,9,8,4)
print(f"total:{total},avg:{avg}")
````

global은 내부와 외부의 객체를 동기화(sync)시켜준다.  
global은 사용하기는 편하지만 언제 바뀌었는지 흐름을 파악하기 힘들기 때문에 남용은 하지 않는다.


````python
c = 3
def proc():
    global c # 외부변수값을 변경하기전에 global로 선언을 해준다.
    c = c+1
    print(c)

proc()

k=1
def outer():
    m = 1
    def inner():
         global k
         k = k + 1
         nonlocal m # 외부 함수의 자원을 내부 함수에서 참조할때는 nonlocal을 사용한다.
         m = m + 1
         return k, m
    return inner()

outer()
````


## 재귀함수


```
재귀함수란?  
  - 함수 내부에서 자기 자신을 다시 호출하는 구조를 가진 함수  
  - 복잡한 문제를 더 작은 단위의 동일한 문제로 나누어 해결할 때 유용 (분할정복)


 재귀함수의 필수 2대 요소  
  - 1. 기저 조건 (Base Case): 재귀 호출을 멈추고 탈출하는 조건 (구현 누락 시 무한 루프)
  - 2. 재귀 단계 (Recursive Case): 문제를 작게 쪼개어 자기 자신을 다시 호출하는 부분


 기본 개념 구조  
   def proc(n):      
        if n <= 1:                    # 1. 기저 조건          
               return 1      
        return n + proc(n-1)     # 2. 재귀 단계 (자기 자신 호출)


```


### 팩토리얼 (n!) 구현
  - n부터 1까지의 모든 양의 정수를 곱하는 연산 (n! = n * (n-1) * (n-2) * ... * 1)

  


### 재귀함수의 치명적인 문제점 3가지
1. 메모리 스택 오버플로우 (Stack Overflow)
  - 함수가 호출될 때마다 메모리의 '스택(Stack)' 영역에 돌아갈 주소와 지역 변수가 쌓임
  - 재귀 호출이 깊어지면 스택 메모리 한도를 초과하여 프로그램이 강제 종료됨


2. 막대한 가독성 및 디버깅 비용
  - 실행 흐름이 직관적이지 않고 복잡하게 얽혀 있어 코드 가독성이 떨어짐
  - 에러 발생 시 추적(Traceback) 과정이 너무 길어져 디버깅(오류 수정)이 까다로움


3. 심각한 중복 연산과 속도 저하
  - 이전에 계산했던 값을 기억하지 못하고 똑같은 연산을 무수히 반복함
  - 숫자가 조금만 커져도 연산량이 기하급수적으로 늘어나 시스템이 멈출 수 있음


````python
def proc_factorial(n):      
    if n <= 1:                    # 1. 기저 조건          
        return 1      
    return n * proc_factorial(n-1)     # 2. 재귀 단계 (자기 자신 호출)

print(proc_factorial(5))
````

### 파이썬(Python) 환경에서의 구체적인 한계


① 파이썬 최대 재귀 깊이 제한 (RecursionError)
  - 파이썬은 스택 오버플로우로 인한 시스템 붕괴를 막는 자체 안전장치가 있음
  - 기본 제한 횟수: 약 1,000회 (이상을 넘어가면 RecursionError 발생)


② 예제 상황 (함수명: proc_factorial)
  - proc_factorial(2000) 호출 시 → 시스템이 버티지 못하고 에러를 뿜으며 멈춤


③꼬리 재귀 최적화(Tail Call Optimization) 미지원
  - 다른 언어(C, Kotlin 등)와 달리 파이썬은 구조적으로 재귀 최적화를 지원하지 않음
  - 즉, 파이썬에서는 대규모 데이터를 다룰 때 재귀함수 사용을 반드시 지양해야 함


### 재귀함수 문제점 해결을 위한 3대 대안


①대안: 반복문(Iteration) 전환
  - for문이나 while문을 사용하면 스택 메모리를 전혀 쓰지 않아 가장 안전하고 빠름


②대안: 메모이제이션(Memoization) 적용
  - 한 번 계산한 결과를 딕셔너리 등에 캐싱(저장)하여 중복 연산 수치를 제로로 만듦


③대안: 동적 계획법(Dynamic Programming) 활용
  - 큰 문제를 작은 하위 문제로 나누되, 아래에서부터(Bottom-up) 상향식으로 채워나감

```
①메모이제이션(Memoization) 기술이란?
  - 컴퓨터 프로그램이 동일한 계산을 반복해야 할 때, 이전에 계산한 값을
    메모리에 저장(캐싱)함으로써 동일한 중복 연산을 제거하는 최적화 기법입니다.


② 핵심 작동 원리 2가지
  - 결과 값 캐싱: 한 번 계산한 결과는 딕셔너리(Dictionary) 등의 메모리 공간에 기록
  - 재귀 호출 스킵: 이미 계산된 값이 필요할 때는 재귀를 수행하지 않고 즉시 꺼내어 재사용


③ 도입 시 기대 효과
  - 무수히 반복되는 중복 연산 수치를 제로(0)에 가깝게 만듦
  - 시간 복잡도를 혁신적으로 줄여 프로그램의 실행 속도를 기하급수적으로 향상시킴
```


````python
# 계산된 결과를 저장할 메모리 공간(딕셔너리) 초기화
memo = {}


def proc_factorial(n):
      # 1. 기저 조건: n이 1 이하일 때 1 반환
      if n <= 1:
          return 1


      # 2. 이미 계산한 적이 있는 값이면 메모리에서 즉시 꺼내어 반환
      if n in memo:
          return memo[n]


      # 3. 계산한 적이 없다면 결과를 계산하여 메모리에 기록(저장) 후 반환
      memo[n] = n * proc_factorial(n - 1)
      return memo[n]

# 함수 호출 및 결과 확인
print(proc_factorial(5))  # 출력: 120
print(memo)               # 출력: {2: 2, 3: 6, 4: 24, 5: 120} (저장된 상태)
````


### 파이썬 파이토닉(Pythonic) 메모이제이션 데코레이터
-   @lru_cache 데코레이터 활용
  - 파이썬 내장 라이브러리를 사용하면 코드 수정 없이 한 줄로 자동 메모리화 가능


````python
from functools import lru_cache


# maxsize=None은 제한 없이 모든 결과를 메모리에 캐싱하겠다는 의미
@lru_cache(maxsize=None)
def proc_factorial(n):
      if n <= 1:
          return 1
      return n * proc_factorial(n - 1)
````
