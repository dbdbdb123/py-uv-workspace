4. positional only
- positional only는 함수를 사용할 때 positional방식으로만 parameter를 쓸 수 있도록 강제하는 방법이다.
- / 의 앞에 오는 parameter들은 positional방식으로 입력할 수 있지만,  keyword방식으로 입력할 수 없다.
- positional only 함수를 정의할 수 있는 것은 python 3.8부터 지원한다. 이전 버전에서는 사전 정의된 positional only 함수는 있지만, 사용자가 만들 수는 없었다.


````python
def proc(a,b,c,/):
    return a,b,c
````

````python
print(proc(4,5,6))
````

````python
print(proc(4,c=6,b=5))
````

````python
#ab는 positional c,d는 positional or keyword
def proc(a,b,/,c,d):
    return a,b,c,d

print(proc(4,5,6,7))
````

5. keyword only
- keyword only방식은 keyword방식으로만 입력하도록 강제하는 방법이다.
- '*' 이후로는 keyword 방식만 가능하다.


````python
def proc(*,a,b,c,d):
    return a,b,c,d

print(proc(a=3,b=3,c=3,d=3))

#print(proc(3,4,c=3,d=4))

def proc(a,b,*,c,d):
    return a,b,c,d

print(proc(3,3,c=3,d=3))
````

6. 가변 positional
- 가변 positional은 parameter 이름에 '*'를 붙인다.
- positional 방식으로 입력되는 argument의 갯수에 상관없이 모두 받아온다.


````python
def proc(*a):
    print(type(a))
    return a

print(proc())
print(proc(1))
print(proc(1,2,3))

# 일반매개변수 뒤에 가변매개변수를 사용하고 가변매개변수를 한번만 사용한다
def proc(*b, a):
    print(type(b))
    print(type(a))
    return a,b


print(proc(1,2,3,4,a=5))
print(proc(1,2,3,4,5))


````

7. 가변 keyword
- 가변 keyword는 parameter 이름에 '**'를 붙인다.
- keyword방식으로 입력되는 argument의 개수에 상관없이 모두 받아온다.


````python
def proc(**a):
    print(type(a))
    return a

proc(x=2,a=1,c=3,d=[1,2,3,4,5,6])
````

````python
def proc(*a,**b):
    return a,b

print(proc(1,2,x=2,y=4))
````

````python
def proc(a,b,/,c,d,*,e,f):
    return a,b,c,d,e,f

print(proc(1,2,3,4,e=5,f=6))
````

## return
  - return의 정의
     - 함수 내부의 실행을 종료하고, 호출한 곳으로 결과값을 전달하는 키워드
  - return의 핵심 역할 2가지  
    - 값 반환: 함수에서 계산되거나 처리된 데이터(결과물)를 외부로 내보냄  
    - 함수 종료: return문이 실행되는 즉시 뒤에 코드가 더 있더라도 함수가 끝남
  - 아무것도 리턴하지  않으면 None을 출력함.  None은 아무것도 없다는 의미임.
  - 함수에서 return을 2번 이상 사용할 수 없다.
  - 함수의 결과를 2개 이상으로 받고 싶다면 unpacking해서 받는다.


````python
def proc(a):
    print(a)
    return # 함수 중간에 빠져나올떄

#아무것도 리턴하지 않으면 none을 출력함. none은 아무것도 없다는 의미
print(proc(3))



````

````python
#proc 함수에서 합계랑 평균을 구해서 리턴하는 프로그램을 구현

def proc(*a):
    total = 0
    avg = 0

    total = sum(a)
    avg = total / len(a)
    return total, avg

total,avg = proc(1,2,3,5,6,7,9,8,4)
print(f"total:{total},avg:{avg}")
````

global은 내부와 외부의 객체를 동기화(sync)시켜준다.  
global은 사용하기는 편하지만 언제 바뀌었는지 흐름을 파악하기 힘들기 때문에 남용은 하지 않는다.


````python
c = 3
def proc():
    global c # 외부변수값을 변경하기전에 global로 선언을 해준다.
    c = c+1
    print(c)

proc()

k=1
def outer():
    m = 1
    def inner():
         global k
         k = k + 1
         nonlocal m # 외부 함수의 자원을 내부 함수에서 참조할때는 nonlocal을 사용한다.
         m = m + 1
         return k, m
    return inner()

outer()
````

## lambda
- python에서는 익명함수를 정의할 때 lambda키워드를 사용한다.
- 간결하게 함수를 표현하거나 def로 정의할 수 없는 경우에 사용한다.
- 형식
  함수이름 = lambda 매개변수1, 매개변수 2,... : 매개변수를 이용한 표현식



````python
c=[1,2,3]
def proc():
    c=[4,5,6] 
    '''
    리스트는 메모리 주소는 스택 값은 힢에 저장됨,
    메모리주소값은 안바뀜 그래서 값 체인지 가능
    '''
    c[0]=50
    print(c)

proc()
````

````python
add = lambda a,b : print(a+b)
add(1,2)
````

함수를 매개변수로 사용하는 내장 함수: map ,filter

````python
# map()함수와의 조합
numbers = (1, 2, 3, 4, 5)
# 리스트의 모든 요소를 하나씩 꺼내어 람다함수로 가공할때 가장 많이 사용
tuple(map(lambda x : x**2, numbers))
````

filter() 함수와의 조합
  - 리스트의 요소 중 람다 함수의 조건식이 참(True)인 요소만 걸러낼 때 사용



````python
numbers = [1,2,3,4,5,6]
proc_dd = filter(lambda x : x %2==0, numbers)
print(proc_dd)
````

sorted() 정렬 기준(key) 지정
-데이터 정렬 시 key 옵션 활용
  - 딕셔너리나 튜플로 이루어진 복잡한 리스트를 정렬할 때 기준점을 지정


````python
students = [('Kim', 85), ('Lee', 92), ('Park', 78)]
proc_storted = sorted(students,key=lambda x: x[1],reverse=True)
print(proc_storted)
````
