Python 기반 오픈소스 웹 애플리케이션 프레임워크로, 데이터 사이언스, 머신러닝, 데이터 분석 결과 등을 쉽고 빠르게 웹 앱 형태로 시각화하고 배포할 수 있도록 설계된 도구. 복잡한 웹 개발 지식 없이도 파이썬 코드만으로 대화형 웹 애플리케이션을 만들 수 있다는 점이 강점




| 주요 특징 | 활용 예시 |
|---|---|
| 파이썬 코드만으로 웹 앱 개발 가능 | 데이터 분석 대시보드 제작 |
| 대화형 위젯(버튼, 슬라이더 등) 제공 | 머신러닝 모델 시연 및 결과 시각화 |
| 빠른 개발 및 배포 지원 | 인터랙티브 데이터 탐색 도구 구현 |
| 데이터 시각화 및 분석에 최적화 | 챗봇 및 대화형 AI 웹앱 개발 |
| 오픈소스, 무료 사용 가능 | 실시간 데이터 모니터링 및 리포트 제공 |
| 별도의 웹 기술(HTML, JS) 불필요 | 사용자 맞춤형 입력 폼 및 데이터 수집 도구 제작 |
