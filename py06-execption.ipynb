```python
class Bird:
    def fly(self):
        raise NotImplemented # 상속받은 자식 클래스에서 fly를 오버라이딩 해줘야 한다


class Eagle(Bird):
    def display(self):
        print('display')

    def fly(self):
        print('very fast')
    

eagle = Eagle()
eagle.fly()
```

```python
from abc import *

class Bird(ABC):
    @abstractmethod
    def fly(self):
        pass

class Eagle(Bird):
    def display(self):
        print('display')

    # def fly(self):
    #     print('very fast')
    
# ABC를 상속받은  bird 추상 클래스. bird을 상속박은 Eagle도 추상 클래스, 추상클래스는 객체 생성을 할수 없다
eagle = Eagle() 
eagle.fly()
```

```python
class UserClass(Exception):
    def __str__(self):
        return '허용되지 않는 데이터 입니다'
```

```python
try:
    x = int(input('숫자 입력:'))
    if x < 0 : 
       raise UserClass
    print(x)
except UserClass as e:
    print(e)
except ValueError as e:
    print(e)
except Exception as e:
    print(e)

"""
구체적
ㅣ
추상적
"""


```
